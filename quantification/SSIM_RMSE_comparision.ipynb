```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
A1 = np.load(path+"/quantification/ssim_rmse_2gens_A1.npy")
A2 = np.load (path+"/quantification/ssim_rmse_2gens_A2.npy")

A1_ssim_vs, A1_ssim_g1_vs, A1_ssim_all, A1_ssim_g1_all, A1_rmse_vs, A1_rmse_g1_vs, A1_rmse_all, A1_rmse_g1_all = A1[0], A1[1], sum(A1[2]), sum(A1[3]), A1[4], A1[5], sum(A1[6]), sum(A1[7])
A2_ssim_vs, A2_ssim_g1_vs, A2_ssim_all, A2_ssim_g1_all, A2_rmse_vs, A2_rmse_g1_vs, A2_rmse_all, A2_rmse_g1_all = A2[0], A2[1], sum(A2[2]), sum(A2[3]), A2[4], A2[5], sum(A2[6]), sum(A2[7])
```

# Plots

```python
plt.rcParams["figure.figsize"] = (11,8)

vs_no = np.arange(len(A1_ssim_vs))+1

fig, ax1 = plt.subplots()
plt.xticks(fontsize=35, rotation=0)


ax1.plot(vs_no, A1_ssim_vs, '-o', color='indigo', label=f'GEN 0, Avg: {A1_ssim_all:.3F}', linewidth= 3,  markersize=7)
ax1.plot(vs_no, A1_ssim_g1_vs, '-', color='dodgerblue', label=f'GEN 1, Avg: {A1_ssim_g1_all:.3F}', linewidth = 4)



ax1.set_xlabel("Video Sequences", size=50,  weight = 'bold')
ax1.set_ylabel("SSIM", size=50, color='tab:blue', weight = 'bold')
ax1.tick_params(axis='y', labelcolor='tab:blue', )
ax1.set_ylim([0.5, 1.09])
plt.yticks(fontsize=34, rotation=0)


ax2 = ax1.twinx()
ax2.plot(vs_no, A1_rmse_vs, '-s', color='red', label=f'GEN 0, Avg: {A1_rmse_all:.2F}',linewidth = 2,)
ax2.plot(vs_no, A1_rmse_g1_vs, '-', color='crimson', label=f'GEN 1, Avg: {A1_rmse_g1_all:.2F}', linewidth=5)
ax2.set_ylabel("RMSE", size=50, color='tab:red', weight = 'bold')
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.set_ylim([-0.10, 1.0])
plt.yticks(fontsize=34, rotation=0)

ax1.legend(loc=(0.01,0.83), prop={'weight':'bold', 'size':'20'}, frameon=False).get_frame().set_facecolor('none')
ax2.legend(loc=(0.45, 0.01), prop={'weight':'bold', 'size':'20'}, frameon=False).get_frame().set_facecolor('none')

ax1.xaxis.set_major_locator(plt.MaxNLocator(4,))
ax1.yaxis.set_major_locator(plt.MaxNLocator(3))
ax2.yaxis.set_major_locator(plt.MaxNLocator(3))

for label in ax1.get_xticklabels(): label.set_fontweight('bold')
for label in ax1.get_yticklabels(): label.set_fontweight('bold')
for label in ax2.get_yticklabels(): label.set_fontweight('bold')    

plt.tight_layout()
# plt.savefig('figures/1I1O_ssim_vs_rmse_GEN1_GEN1.png', transparent=True)
plt.show()
```
