```python
import matplotlib.transforms as transforms
from matplotlib.ticker import MaxNLocator
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
path = os.path.abspath("../..")

A1 = np.load(path+"/Step_4_quantification/Step_3_best_model_selection/ssim_rmse_2gens_A1.npy")
A2 = np.load (path+"/Step_4_quantification/Step_3_best_model_selection/ssim_rmse_2gens_A2.npy")

A1_ssim_vs, A1_ssim_g1_vs, A1_ssim_all, A1_ssim_g1_all, A1_rmse_vs, A1_rmse_g1_vs, A1_rmse_all, A1_rmse_g1_all = A1[0], A1[1], sum(A1[2]), sum(A1[3]), A1[4], A1[5], sum(A1[6]), sum(A1[7])
A2_ssim_vs, A2_ssim_g1_vs, A2_ssim_all, A2_ssim_g1_all, A2_rmse_vs, A2_rmse_g1_vs, A2_rmse_all, A2_rmse_g1_all = A2[0], A2[1], sum(A2[2]), sum(A2[3]), A2[4], A2[5], sum(A2[6]), sum(A2[7])
```

```python
plt.rcParams["figure.figsize"] = (22,8)


fig, (ax3, ax1) = plt.subplots(1, 2,)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=5, hspace=None)

#########################################################
#########################################################

vs_no = np.arange(len(A2_ssim_vs))+1

ax1.plot(vs_no, A2_ssim_vs, '-o', color='darkgreen', label=f'GEN 0, Avg: {A2_ssim_all:.3F}', linewidth= 3,  markersize=7)
ax1.plot(vs_no, A2_ssim_g1_vs, '-', color='c', label=f'GEN 1, Avg: {A2_ssim_g1_all:.3F}', linewidth = 4)

ax1.set_xlabel("Video Sequences", size=50,  weight = 'bold')
ax1.set_ylabel(r"$\mu_{SIM}$", size=80, color='teal', weight = 'bold')
ax1.tick_params(axis='y', labelcolor='teal', labelsize=35)
ax1.tick_params(axis='x', labelcolor='black', labelsize=35)

ax1.set_ylim([0.5, 1.05])

ax2 = ax1.twinx()
ax2.plot(vs_no, A2_rmse_vs, '-s', color='purple', label=f'GEN 0, Avg: {A2_rmse_all:.2F}',linewidth = 2,)
ax2.plot(vs_no, A2_rmse_g1_vs, '-', color='hotpink', label=f'GEN 1, Avg: {A2_rmse_g1_all:.2F}', linewidth=5)
ax2.set_ylabel("RMSE", size=50, color='mediumvioletred', weight = 'bold')
ax2.tick_params(axis='y', labelcolor='mediumvioletred', labelsize=35)
ax2.set_ylim([-0.10, 1.0])

ax1.xaxis.set_major_locator(plt.MaxNLocator(4,))
ax1.yaxis.set_major_locator(plt.MaxNLocator(3))
ax2.yaxis.set_major_locator(plt.MaxNLocator(3))


ax1.legend(loc=(0.01,0.57), prop={'weight':'bold', 'size':'20'},  frameon=False).get_frame().set_facecolor('none')
ax2.legend(loc=(0.45, 0.24), prop={'weight':'bold', 'size':'20'}, frameon=False).get_frame().set_facecolor('none')


for label in ax1.get_xticklabels(): label.set_fontweight('bold')
for label in ax1.get_yticklabels(): label.set_fontweight('bold')
for label in ax2.get_yticklabels(): label.set_fontweight('bold')    

plt.text(-40,1,"(b)", fontsize=70, weight = 'bold')
  
  
#########################################################
#########################################################


ax3.plot(vs_no, A1_ssim_vs, '-o', color='indigo', label=f'GEN-0, Avg: {A1_ssim_all:.3F}', linewidth= 3,  markersize=7)
ax3.plot(vs_no, A1_ssim_g1_vs, '-', color='dodgerblue', label=f'GEN-1, Avg: {A1_ssim_g1_all:.3F}', linewidth = 4)



ax3.set_xlabel("Video Sequences", size=50,  weight = 'bold')
ax3.set_ylabel(r"$\mu_{SIM}$", size=80, color='tab:blue', weight = 'bold')
ax3.tick_params(axis='y', labelcolor='tab:blue', labelsize=35)
ax3.tick_params(axis='x', labelcolor='black', labelsize=35)

ax3.set_ylim([0.5, 1.09])


ax4 = ax3.twinx()
ax4.plot(vs_no, A1_rmse_vs, '-s', color='red', label=f'GEN-0, Avg: {A1_rmse_all:.2F}',linewidth = 2,)
ax4.plot(vs_no, A1_rmse_g1_vs, '-', color='lightcoral', label=f'GEN-1, Avg: {A1_rmse_g1_all:.2F}', linewidth=5)
ax4.set_ylabel("RMSE", size=50, color='tab:red', weight = 'bold')
ax4.tick_params(axis='y', labelcolor='tab:red', labelsize=35)
ax4.set_ylim([-0.10, 1.0])

ax3.legend(loc=(0.01,0.83), prop={'weight':'bold', 'size':'20'}, frameon=False).get_frame().set_facecolor('none')
ax4.legend(loc=(0.45, 0.01), prop={'weight':'bold', 'size':'20'}, frameon=False).get_frame().set_facecolor('none')


ax3.xaxis.set_major_locator(plt.MaxNLocator(4,))
ax3.yaxis.set_major_locator(plt.MaxNLocator(3))
ax4.yaxis.set_major_locator(plt.MaxNLocator(3))

for label in ax3.get_xticklabels(): label.set_fontweight('bold')
for label in ax3.get_yticklabels(): label.set_fontweight('bold')
for label in ax4.get_yticklabels(): label.set_fontweight('bold')    

plt.text(-40,1,"(a)", fontsize=70, weight = 'bold')

plt.tight_layout()
plt.savefig(path+'/Step_4_quantification/Step_3_best_model_selection/best_model.png', transparent=True)

plt.show()
```
