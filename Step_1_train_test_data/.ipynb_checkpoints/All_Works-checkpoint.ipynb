```python
import numpy as np
import pandas as pd
from numpy import savez_compressed
import matplotlib.pyplot as plt
import os
import glob
from natsort import natsorted
from skimage.transform import rescale, resize, downscale_local_mean


os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
os.environ["CUDA_VISIBLE_DEVICES"] = "0" 

from tensorflow.keras.models import load_model
```

## Image Frame Creation from CSV File

```python
model_no = 8

path_M8 = '/home/upadesh/3 Codes/Conv_LSTM/AU_PT_spin_for_lstm/pf_simulations/PF8 Cu-Mn/seed_2/3K/exodus_files/csv_files/'

csv_files = natsorted(glob.glob(path_M8+'*.csv'))

# csv_files = csv_files[50:]

c_values = []
for i, file in enumerate(csv_files):
    df = pd.read_csv(file, index_col=None)
    df = df.sort_values(by =['Points:1', 'Points:0'], ascending = [True, True])
    dist = df['c']
    E = pd.DataFrame(dist)
    E.columns = [os.path.basename(csv_files[i]).split('.')[0]]   ## first it separates whole file path to basename as "00.csv" || and then splits to "00" & "csv" and we need first part "00" only so [0]
    c_values.append(E)
dist = pd.concat(c_values, ignore_index=False, axis=1)

x_y_axis = df[['Points:0', 'Points:1']]
data_M8 = pd.concat([x_y_axis, dist], ignore_index=False, axis=1).iloc[:, 2:]

x_dim = int(data_M8.shape[0]**0.5)
y_dim = int(data_M8.shape[0]**0.5)
t_dim = int(data_M8.shape[1])
print(f'SpatioTemporal(t, x, y) dimension of data is: {t_dim} x {y_dim} x {x_dim}')

M8 = data_M8.to_numpy().reshape(x_dim, y_dim, t_dim)
M8 = np.transpose(M8, (2, 0, 1)) 

np.save('/home/upadesh/3 Codes/Conv_LSTM/AU_PT_spin_for_lstm/pf_simulations/Results/Cu_Mn_M8.npy', M8) 

### Plotting last frame ###
frame_no = t_dim - 1 
plt.imshow(np.squeeze(M8[frame_no]), cmap="rainbow")
plt.title(f'Showing the frame no: {frame_no+1} of Model {model_no}')
plt.gca().invert_yaxis()

M8.shape, data_M8.shape
```

## Video Sequence from Image Frame

```python
path_to_input = '/home/upadesh/3 Codes/Conv_LSTM/AU_PT_spin_for_lstm/pf_simulations/Results/'
path_to_output = '/home/upadesh/3 Codes/Conv_LSTM/AU_PT_spin_for_lstm/pf_simulations/Results/low_high_res/model_2/predictions/Cu-Mn_VS/'

m8 = np.load(path_to_input+'Cu_Mn_M8.npy') 
m8.shape
```

```python
### Lowering Resolution    
def reduce_pixel_size(image, new_shape):
    return resize(image, new_shape, anti_aliasing=True)

new_shape = (m8.shape[0], 64, 64)
m8_low_res = reduce_pixel_size(m8, new_shape)
m8.shape, m8_low_res.shape
```

```python
f, ax = plt.subplot_mosaic('01', figsize=(6, 3))
ax['0'].imshow(m8[-1], cmap="viridis")
ax['0'].invert_yaxis()
ax['0'].set_title("original")

ax['1'].imshow(m8_low_res[-1], cmap="viridis")
ax['1'].invert_yaxis()
ax['1'].set_title("Reduced")
```

```python
##### Low Resolution Images

pfm_datum_low_res = []  # list to store the subarrays
for n in range(1, 16):
    for i in range(0, m8_low_res.shape[0] - 20  , n):
        mini_array_low_res = m8_low_res[i : i + 20, :, :]
        pfm_datum_low_res.append(mini_array_low_res)
        
pfm_datum_low_res = np.array(pfm_datum_low_res)  # convert list to numpy array
np.save(path_to_output+'CuMn.npy', pfm_datum_low_res) 


m8.shape, pfm_datum_low_res.shape, # pfm_datum.shape,
```

```python
# Construct a figure on which we will visualize the images.
fig, axes = plt.subplots(4, 5, figsize=(8, 6))

# Plot each of the sequential images for one random data example.
data_choice = np.random.choice(range(len(pfm_datum_low_res)), size=1)[0]
for idx, ax in enumerate(axes.flat):
    ax.imshow(np.squeeze(pfm_datum_low_res[data_choice][idx]), cmap="viridis")
    ax.set_title(f"Frame {idx + 1}")
    ax.axis("off")

# Print information and display the figure.
print(f"Displaying frames for example {data_choice}.")
plt.show()
```

## Prediction

```python
path = os.path.abspath("..")

X_val = np.load(path+"/Cu-Mn_VS/X.npy")
y_val = np.load (path+"/Cu-Mn_VS/y.npy")

X_val.shape, y_val.shape
```

```python
model = load_model(path+"/trained_model.h5")
history = np.load(path+"/model_history.npy", allow_pickle=True).item()
```

```python
def scale_array(arr):
    arr_min = np.min(arr)  ## Flattens the array and finds the min value inside it
    arr_max = np.max(arr)  ## Flattens the array and finds the max value inside it
    scaled_array = (arr - arr_min) / (arr_max - arr_min) * 255
    scaled_array = scaled_array.astype(np.uint8)
    
    return scaled_array
```

```python
GT_array = y_val
a, b, c, d, e = y_val.shape[0], y_val.shape[1], y_val.shape[2], y_val.shape[3], y_val.shape[4] 
PD_array = np.zeros((a, b, c, d, e))   ## Predicted arrays

for i in range(X_val.shape[0]):
    frames = X_val[i]   ## taking 1 validation video sample at one time
    PD_array[i] = model.predict(np.expand_dims(frames, axis=0), verbose=0, workers=10, use_multiprocessing=True)

GT_scaled = scale_array(y_val)
PD_scaled = scale_array(PD_array)

GT_array.shape, PD_array.shape
```

```python
X_g1 = np.delete(np.concatenate((X_val, PD_array), axis=-1), 0, axis=-1)
X_g1 = np.delete(X_g1, -1, axis=1)
y_g1 = np.delete(y_val, 0, axis=1)

a, b, c, d, e = y_g1.shape[0], y_g1.shape[1], y_g1.shape[2], y_g1.shape[3], y_g1.shape[4] 
PD_G1 = np.zeros((a, b, c, d, e))   ## Predicted arrays
for i in range(X_g1.shape[0]):
    frames = X_g1[i]   ## taking 1 validation video sample at one time
    PD_G1[i] = model.predict(np.expand_dims(frames, axis=0), verbose=0, workers=10, use_multiprocessing=True)

y_g1_scaled = scale_array(y_g1)
PD_G1_scaled = scale_array(PD_G1)

y_g1.shape, PD_G1.shape
```

```python
X_g2 = np.delete(np.concatenate((X_g1, PD_G1), axis=-1), 0, axis=-1)
X_g2 = np.delete(X_g2, -1, axis=1)
y_g2 = np.delete(y_g1, 0, axis=1)

a, b, c, d, e = y_g2.shape[0], y_g2.shape[1], y_g2.shape[2], y_g2.shape[3], y_g2.shape[4] 
PD_G2 = np.zeros((a, b, c, d, e))   ## Predicted arrays

for i in range(X_g1.shape[0]):
    frames = X_g2[i]   ## taking 1 validation video sample at one time
    PD_G2[i] = model.predict(np.expand_dims(frames, axis=0), verbose=0, workers=10, use_multiprocessing=True)

y_g2_scaled = scale_array(y_g2)
PD_G2_scaled = scale_array(PD_G2)

y_g2.shape, PD_G2.shape
```

```python
X_g3 = np.delete(np.concatenate((X_g2, PD_G2), axis=-1), 0, axis=-1)
X_g3 = np.delete(X_g3, -1, axis=1)
y_g3 = np.delete(y_g2, 0, axis=1)

a, b, c, d, e = y_g3.shape[0], y_g3.shape[1], y_g3.shape[2], y_g3.shape[3], y_g3.shape[4] 
PD_G3 = np.zeros((a, b, c, d, e))   ## Predicted arrays

for i in range(X_g2.shape[0]):
    frames = X_g3[i]   ## taking 1 validation video sample at one time
    PD_G3[i] = model.predict(np.expand_dims(frames, axis=0), verbose=0, workers=10, use_multiprocessing=True)

y_g3_scaled = scale_array(y_g3)
PD_G3_scaled = scale_array(PD_G3)

y_g3.shape, PD_G3.shape
```

```python
np.save(path+'/Cu-Mn_VS/GEN-0.npy', [GT_array, PD_array])
np.save(path+'/Cu-Mn_VS/GEN-1.npy', [y_g1, PD_G1])
np.save(path+'/Cu-Mn_VS/GEN-2.npy', [y_g2, PD_G2])
np.save(path+'/Cu-Mn_VS/GEN-3.npy', [y_g3, PD_G3])

# np.save(path+'/Cu-Mn_VS/for_blobs_shape_index/GEN-0.npy', [GT_array, PD_array])
# np.save(path+'/Cu-Mn_VS/for_blobs_shape_index/GEN-1.npy', [y_g1, PD_G1])
# np.save(path+'/Cu-Mn_VS/for_blobs_shape_index/GEN-2.npy', [y_g2, PD_G2])
# np.save(path+'/Cu-Mn_VS/for_blobs_shape_index/GEN-3.npy', [y_g3, PD_G3])
```
